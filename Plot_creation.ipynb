```python
def calculate_step_size(data_range):
    if data_range <= 1:
        return 0.2
    elif data_range <= 3:
        return 0.6
    elif data_range <= 10:
        return 1.5
    elif data_range <= 50:
        return 10
    elif data_range <= 100:
        return 20
    elif data_range <= 300:
        return 50
    else:
        return 100
```

```python
# Now load the two text files with the slope and intercept values
# and save them to a DataFrame

import pandas as pd

# Read the text file
with open('Slope_intercept_rtm_only.txt', 'r') as file:
    lines = file.readlines()

# Initialize lists to store parsed values
slope = []
intercept = []
variable = []
rse = []

# Parse the content and extract values
for i in range(0, len(lines), 5):
    slope_val = float(lines[i].split(':')[1].strip())
    intercept_val = float(lines[i + 1].split(':')[1].strip())
    rse_val = float(lines[i + 2].split(':')[1].strip())
    variable_val = lines[i + 3].split(':')[1].strip()
    
    slope.append(slope_val)
    intercept.append(intercept_val)
    variable.append(variable_val)
    rse.append(rse_val)

# Create a DataFrame
data = {
    'Slope': slope,
    'Intercept': intercept,
    'RSE': rse,
    'Variable': variable
}

df_rtm = pd.DataFrame(data)



# Read the text file
with open('Slope_intercept_real_only.txt', 'r') as file:
    lines = file.readlines()

# Initialize lists to store parsed values
slope = []
intercept = []
variable = []
rse = []

# Parse the content and extract values
for i in range(0, len(lines), 5):
    slope_val = float(lines[i].split(':')[1].strip())
    intercept_val = float(lines[i + 1].split(':')[1].strip())
    rse_val = float(lines[i + 2].split(':')[1].strip())
    variable_val = lines[i + 3].split(':')[1].strip()
    
    slope.append(slope_val)
    intercept.append(intercept_val)
    variable.append(variable_val)
    rse.append(rse_val)

# Create a DataFrame
data = {
    'Slope': slope,
    'Intercept': intercept,
    'RSE': rse,
    'Variable': variable
}

df_real_only = pd.DataFrame(data)

```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import os

# Load all plots that have a name starting with "Scatterplot" and ending with ".png"
files = os.listdir()
files = [file for file in files if file.startswith('Scatterplot') and file.endswith('.png')]
files.sort()

# Set the number of columns and rows
cols = 4
rows = 4

# Set the figure size
plt.rcParams['figure.figsize'] = [20, 20]

# Plot the subplots alphabetically in a grid with no space between the plots
for i, file in enumerate(files):
    plt.subplot(rows, cols, i + 1)
    img = mpimg.imread(file)
    plt.imshow(img)
    plt.axis('off')

# Add a centered legend between two empty subplots including the explanatory text
plt.subplot(rows, cols, (rows * cols) - 1)  # Select the penultimate subplot
plt.axis('off')  # Turn off axis for the penultimate subplot

# Define the legend items with customized markers and sizes
legend_items = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=15, label='Real Data Only'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=15, label='RTM Data Only')
]

# Create the legend with frames and include the explanatory text
plt.legend(handles=legend_items, loc='center', bbox_to_anchor=(1, 0.65), frameon=True, prop={'size': 15}, ncol=2,
           title='Legend', title_fontsize='large')
#plt.text(0.3, 0.5, 'X-axis = Predictions', horizontalalignment='left',
 #        verticalalignment='center', transform=plt.gca().transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))
#plt.text(1.1, 0.5, 'Y-axis = Observations', horizontalalignment='left',
 #        verticalalignment='center', transform=plt.gca().transAxes, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

# Save the figure
plt.subplots_adjust(wspace=0, hspace=0)  # Ensure no space between subplots
plt.savefig('All_scatterplots_with_legend_and_text_rtm_only.png', bbox_inches='tight')
# Save figure as pdf
plt.savefig('All_scatterplots_with_legend_and_text_rtm_only.pdf', bbox_inches='tight')
plt.show()
```

```python
#### REST IS DEPRECATED ####





# Load all the png plots in the plot_folder and plot a maximum of two plots in a 1x2 grid for each number in the filename
import os
from PIL import Image
import matplotlib.pyplot as plt

# Define the folder path where your PNG files are located
folder_path = '/net/home/dmederer/Transferability_efficient/plot_folder'

# List of numbers you mentioned
numbers = [10, 11, 12, 14, 15, 16, 21]

# Create a dictionary to group images by number
image_groups = {num: [] for num in numbers}

# Loop through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.png'):
        for num in numbers:
            if str(num) in filename:
                img = Image.open(os.path.join(folder_path, filename))
                if len(image_groups[num]) < 2:  # Limit to two images per number
                    image_groups[num].append((filename, img))

# Sort images based on filenames
for num, images in image_groups.items():
    image_groups[num] = sorted(images, key=lambda x: x[0])[:2]  # Sort based on filenames and limit to 2 images

# Create subplots for each group of images and save as PNG
for num, images in image_groups.items():
    if images:
        num_images = len(images)
        num_cols = 2
        num_rows = 1  # Fixed to 1 row since we're allowing a maximum of 2 images
        
        #fig = plt.subplots(num_rows, num_cols, figsize=(30, 15), gridspec_kw={'wspace': 0.000001})
        fig = plt.figure(figsize=(10,5)) # Notice the equal aspect ratio
        axs = [fig.add_subplot(1,2,i+1) for i in range(2)]

        for a in axs:
            a.set_xticklabels([])
            a.set_yticklabels([])
            a.set_aspect('equal')
        # Adjust figure size as needed
        #fig, axs = plt.subplots(num_rows, num_cols, figsize=(30, 5), gridspec_kw={'wspace': 0.000001})

        # fig.suptitle(f'Images with Number {num}')

        for i, (filename, img) in enumerate(images):
            row = 0
            col = i % num_cols
            axs[col].imshow(img)
            # axs[col].set_title(f'Image {i + 1}')
            axs[col].axis('off')

        # Hide any empty subplot
        for i in range(num_images, num_cols):
            axs[i].axis('off')

        #plt.tight_layout()
        #plt.subplots_adjust(top=0.9)
        # Manually adjust spacing between subplots to avoid padding
        plt.subplots_adjust(wspace=0, hspace=0)  # Adjust these values as needed


        # Save the 1x2 grid as a PNG file
        save_path = f'Comparison_big_model_with_rtm_total_1x2_{num}.png'
        plt.savefig(save_path)
        plt.close()

print("Grids saved as PNG files.")
```

```python
## Combine comparison plots into one image

from PIL import Image

# File names of your PNG images
file_names = [
    'Comparison_big_model_with_rtm_total_1x2_10.png',
    'Comparison_big_model_with_rtm_total_1x2_11.png',
    'Comparison_big_model_with_rtm_total_1x2_12.png',
    'Comparison_big_model_with_rtm_total_1x2_14.png',
    'Comparison_big_model_with_rtm_total_1x2_15.png',
    'Comparison_big_model_with_rtm_total_1x2_16.png',
    'Comparison_big_model_with_rtm_total_1x2_21.png'
]

# Load images and sort them based on the numeric part of the file names
images = [Image.open(filename) for filename in file_names]
images.sort(key=lambda x: int(x.filename.split('_')[-1].split('.')[0]))

for img in images:
    print(img.size)

# Split the images into two groups based on the layout order
first_two_vertically = images[:2]
next_two_horizontally = images[2:4]
next_two_vertically = images[4:6]
last_image = images[6]

# Get dimensions of the images
widths_first_two_vertically, heights_first_two_vertically = zip(*(i.size for i in first_two_vertically))
widths_next_two_horizontally, heights_next_two_horizontally = zip(*(i.size for i in next_two_horizontally))
widths_next_two_vertically, heights_next_two_vertically = zip(*(i.size for i in next_two_vertically))
width_last_image, height_last_image = last_image.size

# Calculate the total width and height for the final image
max_width = sum([max(widths_first_two_vertically), max(widths_next_two_horizontally)])
total_height = sum(heights_first_two_vertically) + sum(heights_next_two_vertically)

# Create a new blank image with the calculated dimensions
merged_image = Image.new('RGB', (max_width, total_height), color='white')

# Paste images in the specified layout
y_offset = 0
for img in first_two_vertically:
    merged_image.paste(img, (0, y_offset))
    y_offset += img.height

x_offset = max(widths_first_two_vertically)
y_offset = 0
for img in next_two_horizontally:
    merged_image.paste(img, (x_offset, y_offset))
    y_offset += img.height

y_offset = sum(heights_first_two_vertically)
for img in next_two_vertically:
    merged_image.paste(img, (0, y_offset))
    y_offset += img.height

x_offset = max(widths_next_two_horizontally)
y_offset = sum(heights_first_two_vertically)
merged_image.paste(last_image, (x_offset, y_offset))

# Save the merged image
merged_image.save('big_model_merged_image.png')
```
